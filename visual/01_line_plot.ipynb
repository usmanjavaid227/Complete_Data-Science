## adding limits

```python

import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
phool=sns.load_dataset("iris")
#phool
#draw a linr plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("phoolon ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
phool=sns.load_dataset("iris")
#phool
#draw a linr plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("phoolon ka plot")
sns.set_theme()
sns.set_style("dark")
plt.show
```

## size of figure

```python

import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
phool=sns.load_dataset("iris")
#phool

# set size
plt.figure(figsize=(10,8))
#draw a linr plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("phoolon ka plot")

plt.show
```
