```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

nukta=sns.load_dataset("dots")
pelette= sns.color_palette("pastel")
sns.relplot(data=nukta,x="time",y="firing_rate",size="choice",col="align",kind="line",palette=pelette,
                    height=5,aspect=0.80,facet_kws=dict(sharex=False))
```
