# Read CSV files

```python
import pandas as pd
a=pd.read_csv("data_viz.csv")
a
```

##### Show specific no. of row

```python
a=pd.read_csv("data_viz.csv",nrows=3)
a
```

##### To see any column

```python
a=pd.read_csv("data_viz.csv",usecols=[0,1])
a
```

#### If you want to skip a ROw

```python
a=pd.read_csv("data_viz.csv",skiprows=[0])
a
```

##### Change Index

```python
a=pd.read_csv("data_viz.csv",index_col="Gender")
a
```
