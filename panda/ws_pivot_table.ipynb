```python
import pandas as pd
```

```python
var=pd.DataFrame({
    'days':[1,2,3,4,5,6],
    'eng':[10,20,30,40,50,60],
    'math':[20,30,40,50,60,70]
})
var
```

```python
# to reshape this 
pd.melt(var)
```

```python
# to reshape this 
pd.melt(var,id_vars=['days'])
```

```python

pd.melt(var,id_vars=['days'],var_name='python')
```

# pivot

```python
v=pd.DataFrame({
    'days':[1,2,3,4,5,6],
    "st_name":['a','b','c','a','b','c'],
    'eng':[10,20,30,40,50,60],
    'math':[20,30,40,50,60,70]
})
v
```

```python
v.pivot(index='days',columns='st_name')
```

```python
v.pivot(index='days',columns='st_name',values='eng')
```

```python
v.pivot_table(index="st_name",columns="days",aggfunc='sum',margins=True)
```
