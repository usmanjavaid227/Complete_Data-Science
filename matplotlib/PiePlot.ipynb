# _**PiePlot**_

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['C','C++',"java","python"]
```

```python
plt.pie(x)
plt.show()
```

```python
plt.pie(x,labels=y)
plt.show()
```

```python
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
c=['r','g','b','y']
plt.pie(x,labels=y,colors=c,autopct="%0.1f%%")
plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.1f%%",shadow=True,radius=1.2)
plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.1f%%",shadow=True,radius=1.2,labeldistance=1.1,explode=ex)
plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.1f%%",shadow=True,radius=1.2,labeldistance=1.1,explode=ex,startangle=180,textprops={'fontsize':12})
plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.1f%%",counterclock=False,explode=ex,shadow=True,wedgeprops={"linewidth":5,'edgecolor':'b'})
plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.1f%%",center=(1,5),explode=ex)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,rotatelabels=True)
plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.1f%%",shadow=True,explode=ex)
plt.title("WScube")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x1=[40,30,20,10]
c=['r','y','b','g']
plt.pie(x,labels=y,radius=1,colors=c)
plt.pie(x1,radius=0.5)
plt.show()
```

```python
c=['r','y','b','g']
plt.pie(x,labels=y,radius=1.5,colors=c)
plt.pie([1],colors='w')
plt.show()
```

```python
plt.pie(x,labels=y,radius=1.5,colors=c)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
