# Matplotlib _**BarPlot**_

```python
# import libraries
import matplotlib.pyplot as plt
```

```python
# define data
x=['python','c','c++','java']
y=[85,70,60,82]
```

```python
# plot the bar chart
plt.bar(x,y)
# show labels on x-axis & y-axis 
plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity No.",fontsize=10)
# show title
plt.title("language table",fontsize=15)
plt.show()
```

```python
# define bar's width
plt.bar(x,y,width=0.4)
plt.show()
```

```python
# change color of bar
plt.bar(x,y,color="blue")
plt.show()
```

```python
# change color of bar of different group of colors
c=['y','b','m','g']
plt.bar(x,y,color=c)
plt.show()
```

```python
# change to label position ---> by default it is in center
plt.bar(x,y,align='edge')
plt.show()
```

```python
# change style , color of bar , width of bar, alpha of bar(transparency)
plt.bar(x,y,edgecolor='b',linewidth=3,linestyle=':',alpha=0.8)
plt.show()
```

```python
# you can define legend of bar
plt.bar(x,y,color='r',label="popularity")
# show legend
plt.legend()
plt.show()
```

```python
# two different values for axis on same graph
z=[22,43,14,50]
plt.bar(x,y,color='r',label="popularity")
plt.bar(x,z,color='g',label="popularity1")
plt.legend()
plt.show()
```

If you to show the different values of an axis, you can do this by using numpy libraries by defining differe width of bars.

```python
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]

 # define initial bar's width
width=0.4
 # different width for p and p1
p=np.arange(len(x))
p1=[j+width for j in p]


plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity No.",fontsize=10)
plt.title("language table",fontsize=15)

plt.bar(p,y,width,color='r',label="popularity")
plt.bar(p1,z,width,color='y',label="popularity1")
plt.xticks(p+width/2,x,rotation=25)

plt.legend(loc="upper left",)
plt.show()
```

## How to draw barplot Horizentally ?

```python
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity No.",fontsize=10)
plt.title("language table",fontsize=15)

plt.barh(x,y,color='r',label="popularity")
plt.barh(x,z,color='g',label="popularity1")

plt.legend()
plt.show()
```
