# Matplotlib _**Bar Plot**_

```python
import matplotlib.pyplot as plt
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
```

```python
plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity No.",fontsize=10)
plt.title("language table",fontsize=15)
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,width=0.4)
plt.show()
```

```python
plt.bar(x,y,color="blue")
plt.show()
```

```python
c=['y','b','m','g']
plt.bar(x,y,color=c)
plt.show()
```

```python
plt.bar(x,y,align='edge')
plt.show()
```

```python
plt.bar(x,y,edgecolor='m',linewidth=4,linestyle=':',alpha=0.6)
plt.show()
```

```python
plt.bar(x,y,color='r',label="popularity")
plt.legend()
plt.show()
```

```python
z=[22,43,14,50]
plt.bar(x,y,color='r',label="popularity")
plt.bar(x,z,color='g',label="popularity1")
plt.legend()
plt.show()
```

```python
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]


plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity No.",fontsize=10)
plt.title("language table",fontsize=15)

plt.bar(p,y,width,color='r',label="popularity")
plt.bar(p1,z,width,color='y',label="popularity1")
plt.xticks(p+width/2,x,rotation=25)

plt.legend(loc="upper left",)
plt.show()
```
