# **_StemPlot_**

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,2,1,4,3,3]
```

```python
plt.stem(x,y)
plt.show()
```

```python
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
plt.stem(x,y,linefmt=":",markerfmt='r+')
plt.show()
```

```python
plt.stem(x,y,bottom=3,basefmt='y')
plt.show()
```

```python
plt.stem(x,y,label="python")
plt.legend()
plt.show()
```

```python
plt.stem(x,y,linefmt=":",markerfmt='r+',bottom=0,basefmt='y',use_line_collection=False,orientation="horizontal")
plt.show()
```
