# _**ScatterPlot**_

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,4,1,3,4,5]
```

```python
plt.scatter(day,no)
plt.show()
```

```python
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no)
plt.show()
```

```python
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no,color="r")
plt.show()
```

```python
colors=['r','m','y','g','b','m','r']
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no,c=colors)
plt.show()
```

```python
colors=['r','m','y','g','b','m','r']
sizes=[680,442,974,405,102,790,1017]
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5)
plt.show()
```

```python
colors=['r','m','y','g','b','m','r']
sizes=[680,442,974,405,102,790,1017]
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no,c=colors,s=sizes,alpha=0.7,marker="*",edgecolor="r",linewidth=2)
plt.show()
```

```python
colors=[10,20,30,67,90,35,80]
sizes=[680,442,974,405,102,790,1017]
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no,c=colors,s=sizes,cmap="BrBG")
t=plt.colorbar()
t.set_label("colorbar")
plt.show()
```

```python
colors=[10,20,30,67,90,35,80]
no2=[1,4,9,4,5,1,7]
sizes=[680,442,974,405,102,790,1017]
plt.xlabel("Day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.title("scatter plot",fontsize=15)
plt.scatter(day,no,c=colors,s=sizes,cmap="BrBG")
plt.scatter(day,no2,c='r',s=sizes)
t=plt.colorbar()
t.set_label("colorbar")
plt.show()
```
