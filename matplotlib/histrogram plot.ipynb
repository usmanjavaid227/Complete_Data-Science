# **_Hoistram_** _Plot_

```python
import matplotlib.pyplot as plt 
import numpy as np
import random
```

```python
x=np.random.randint(10,60,(50))
x
```

```python
no=[28, 32, 14, 18, 22, 32, 52, 23, 54, 26, 17, 38, 38, 34, 29, 14, 39,
       22, 33, 31, 29, 36, 26, 56, 12, 30, 23, 35, 43, 17, 15, 26, 10, 13,
       50, 46, 42, 25, 41, 43, 14, 36, 17, 45, 15, 22, 49, 31, 22, 56]
```

```python
plt.hist(no)
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python
plt.hist(no,color="g",edgecolor="r")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python
l=[10,20,30,40,50,60]
plt.hist(no,color="g",bins=l,edgecolor="r")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python
l=[10,20,30,40,50,60]
plt.hist(no,bins='auto',range=(0,100),edgecolor="r")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python
plt.hist(no,color="g",cumulative=-1,edgecolor="r")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python
plt.hist(no,color="g",bottom=10,edgecolor="r")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python
plt.hist(no,color="g",align="left",edgecolor="r")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python

plt.hist(no,color="g",edgecolor="r",histtype="step")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python

plt.hist(no,color="g",edgecolor="r",orientation="horizontal",rwidth=0.7)
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python

plt.hist(no,color="g",edgecolor="r",log=True,rwidth=0.9)
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.show()
```

```python

plt.hist(no,color="g",edgecolor="r",label="python")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.legend()
plt.show()
```

```python

plt.hist(no,color="g",edgecolor="r",label="python")
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.axvline(40,color="r",label="Line")
plt.legend()
plt.grid()
plt.show()
```
