# **Fill_Between Plot**

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,5,4,3,4]
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g')
plt.show()
```

# **SubPlot**

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

plt.subplot(2,2,2)
plt.pie([1],colors="r")

plt.subplot(2,2,3)
plt.bar(x,y)

plt.subplot(2,2,4)
plt.pie(y)

plt.show()
```

# **savePlot as png**

```python
x=[1,2,3,4,5]
y=[3,5,4,3,4]
plt.plot(x,y,color='r')
plt.savefig('line',dpi=5000,facecolor='y')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,5,4,3,4]
plt.plot(x,y,color='r')
plt.savefig('line1.pdf',dpi=1000,facecolor='y')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,5,4,3,4]
plt.plot(x,y,color='r')
plt.savefig('line3',dpi=1000,transparent=True)
plt.show()
```
