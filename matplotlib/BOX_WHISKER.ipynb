# **_BOX_WHISKER_Plot_**

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.boxplot(x,notch=True)
plt.show()
```

```python
plt.boxplot(x,vert=False,widths=0.5)
plt.show()
```

```python
plt.boxplot(x,labels=['python'],patch_artist=True)
plt.show()
```

```python
plt.boxplot(x,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x)
plt.show()
```

```python
plt.boxplot(x,whis=2)
plt.show()
```

```python
plt.boxplot(x,sym='g+',boxprops=dict(color='r'),capprops=dict(color='b'),whiskerprops=dict(color='r'))
plt.show()
```
